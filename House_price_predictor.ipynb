```python
import pandas as pd
import numpy as np
df = pd.read_csv("/content/sample_data/california_housing_train.csv")

test_df = pd.read_csv("/content/sample_data/california_housing_test.csv")
df["median_house_value"] = df["median_house_value"].astype(int)
df
```

```python
from sklearn.linear_model import LinearRegression
regr = LinearRegression()
features = df[["housing_median_age", "total_rooms", "total_bedrooms", "households", "median_income", "longitude", "latitude"]]
y_vals = df[["median_house_value"]]
regr.fit(features, y_vals)
```

```python
test_df
features2 = test_df[["housing_median_age", "total_rooms", "total_bedrooms", "households", "median_income", "longitude", "latitude"]]
y_vals2 = test_df[["median_house_value"]]
```

```python
predictions = regr.predict(features2)
```

```python
test_df["predictions"] = predictions
```

```python
test_df
```

```python
regr.score(features2, y_vals2)
```

```python
import matplotlib.pyplot as plt
plt.plot(test_df.index[:100], test_df.median_house_value[:100])
plt.plot(test_df.index[:100], predictions[:100])
plt.show
```
